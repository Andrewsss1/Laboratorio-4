```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
ang = [48, 50, 52, 54, 56, 58, 60, 62, 64, 66, 68, 70, 72, 74, 76, 78, 80, 82, 84, 86, 88]
pot = [7, 18, 70, 167, 295, 450, 555, 685, 774, 875, 964, 975, 971, 905, 856, 701, 482, 255, 82, 15, 6]
```

```python
ang = np.array(ang)
ang_rel = ang - 68
print(ang_rel)
pot = np.array(pot)
ang_rad = ang_rel * pi /180
print(ang_rad)
```

```python
sin_ang = np.sin(ang_rad)
print(sin_ang)
```

```python
plt.figure(figsize = (10,10))
plt.scatter( sin_ang, pot)
plt.yscale("log")
plt.show()
```

```python
max_pow = max(pot)
pot_perc = pot * 100/ max_pow
```

```python
print(pot_perc)
```

```python
plt.figure(figsize = (10,10))
plt.scatter( ang_rad, pot_perc)
plt.plot([-0.3, 0.3], [5,5], color = "red")
plt.yscale("log")
plt.show()
```
