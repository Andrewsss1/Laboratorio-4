```python
import ROOT
import pandas as pd
import Functions
import numpy as np
import array
import matplotlib.pyplot as plt
from math import e
```

```python
Iname_drift = []
for i in range(305,329):
    Iname_drift.append("PICO_I_run0" + str(i) + ".csv")
for i in range(330,339):
    Iname_drift.append("PICO_I_run0" + str(i) + ".csv")
```

```python
colonne = ["I3B", "I3T", "I2B","I2T", "I1B", "I1T","ID"]
f1 = ROOT.TF1( "f", "gaus",-60,0)
folder = "arrays_corr/pico_data"
```

```python
c = list(np.zeros(7))
j = 0
H = list(np.zeros(7))
ParI_drift = []
SigI_drift = []

for i in colonne:
    c[j] = ROOT.TCanvas("", "", 3000, 6000)
    c[j].Divide(5,7)
    c[j].Draw()
    H[j], parI_drift, sigI_drift =Functions.Do_histo(folder,Iname_drift, f1, i,"I3B", 60, c[j])
    ParI_drift.append(parI_drift)
    SigI_drift.append(sigI_drift)
    for k in range(len(Iname_drift)):
        H[j][k].SetLineColor(j*j+50)
        H[j][k].Draw()
    j+= 1
```

```python
Vname_drift = []
for i in range(305,329):
    Vname_drift.append("PICO_V_run0" + str(i)+ ".csv")
for i in range(330,339):
    Vname_drift.append("PICO_V_run0" + str(i)+ ".csv")
parV_drift,sigV_drift = Functions.V_mean(folder, Vname_drift,6)
parV_drift = np.array(parV_drift)/3
sigV_drift = np.array(sigV_drift)/3
```

```python
c1= ROOT.TCanvas("","",700,700)
graph = list(np.zeros(7))
graph_err = list(np.zeros(7))
j=0
for i in range(7):
    graph[i] = ROOT.TGraph(len(ParI_drift[i]),array.array("f",parV_drift), array.array("f",abs(np.array(ParI_drift[i]))))
    graph_err[i] = ROOT.TGraphErrors(len(ParI_drift[i]), array.array("f", parV_drift),
                                         array.array("f", abs(np.array(ParI_drift[i]))), array.array("f", sigV_drift),
                                         array.array("f", SigI_drift[i]))
    graph[i].SetMarkerStyle(i+2)
    graph[i].SetMarkerColor(i*i + 50)
    if i == 0:
        graph[i].Draw("AP")
        graph[i].SetTitle("I vs E_{drift}")
        graph[i].GetXaxis().SetTitle("E_{drift}")
        graph[i].GetYaxis().SetTitle("I (nA)")
    else:
        graph[i].Draw("SAMEP")
    graph_err[i].Draw("SAMEP")
c1.Draw()
c1.SetLogy()


Leg = ROOT.TLegend(0.1,0.7,0.2,0.9)
Leg.SetHeader("Legend","C")
for i in range(6):
    Leg.AddEntry(graph[i], colonne[i])
Leg.Draw()
```

```python
plt.figure(figsize = (10,10))
for i in range(len(ParI_drift)):
    #plt.scatter(parV_drift, ParI_drift[i], label = colonne[i])
    plt.errorbar(parV_drift, ParI_drift[i], xerr = sigV_drift, yerr = SigI_drift[i], label = colonne[i], linestyle = " ", marker = "o")
#plt.yscale("log")
plt.ylabel("I (nA)")
plt.title("$I \, vs \, E_{drift}$")
plt.xlabel("$E_{drift} (kV/m)$")

plt.legend()
plt.show()
```
