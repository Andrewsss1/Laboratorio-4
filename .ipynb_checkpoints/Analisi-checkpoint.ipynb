# Andamento delle prime 1005 misure

```python
f = open("dati/400ms_3h_data2.txt", "r")
lines = f.read().split("\n")
```

```python
def Create_histo(var, name, bins, x0,x1):
    h = ROOT.TH1F(name, "", bins, x0,x1)
    for i in var:
        h.Fill(i)
    return h
```

```python
T_raw=[]
T_c=[]
U_raw=[]
U_p=[]

for i in range(2, 3015, 3):

    t_raw = lines[i].split("  ")[10]
    t_c = lines[i].split("  ")[11]
    u_raw = lines[i].split("  ")[12]
    u_p = lines[i].split("  ")[13]
    T_raw.append(float(t_raw))
    T_c.append(float(t_c))
    U_raw.append(float(u_raw))
    U_p.append(float(u_p))
```

```python
c1 = ROOT.TCanvas("", "", 1200,700)
g1_1 = ROOT.TGraph(len(T_c), array.array("f",T_c), array.array("f",U_p))
g1_1.SetMarkerStyle(3)
g1_1.SetMarkerColor(98)
g1_1.SetLineColor(98)
g1_1.GetXaxis().SetTitle("T(^Â°C)")
g1_1.GetYaxis().SetTitle("RH(%)")
g1_1.SetTitle("Grafico RH vs T prime 1000 misure")
g1_1.Draw("APL")
c1.Draw()
```

## Analisi per il coefficiente di correlazione
E' stata fatta usando come errore
$$ \Delta \rho = \frac {1-\rho^2}{\sqrt{N-2}}$$

```python
rho_5 = []
rho_10=[]
rho_15 = []
rho_err_5 =[]
rho_err_10 = []
rho_err_15 = []

for i in range(0, len(Tc_5)):
    rho = np.corrcoef(Tc_5[i], Up_5[i])
    #print(rho)
    if pd.isna(rho[0,1]) == True:
        rho[0,1] = 0
    rho_5.append(rho[0,1])
    rhoerr = (1- rho[0,1]**2)/np.sqrt(5) 
    rho_err_5.append(rhoerr)


for i in range(0, len(Tc_10)):
    rho = np.corrcoef(Tc_10[i], Up_10[i])
    #print(rho)
    if pd.isna(rho[0,1]) == True:
        rho[0,1] = 0
    rho_10.append(rho[0,1])
    rhoerr = (1- rho[0,1]**2)/np.sqrt(8) 
    rho_err_10.append(rhoerr)


    
for i in range(0, len(Tc_15)):
    rho = np.corrcoef(Tc_15[i], Up_15[i])
    #print(rho)
    if pd.isna(rho[0,1]) == True:
        rho[0,1] = 0
    rho_15.append(rho[0,1])
    rhoerr = (1- rho[0,1]**2)/np.sqrt(13) 
    rho_err_15.append(rhoerr)


rho_mean_5 = np.mean(rho_5)
rho_mean_10= np.mean(rho_10)
rho_mean_15 = np.mean(rho_15)
```

```python
c = ROOT.TCanvas("", "a values", 1200,900)
c.Divide(2,3)

c.cd(1)
h1 = Create_histo(rho_5, "rho_val", 15, -1.5,1.5)
h1.SetLineColor(1)
h1.SetTitle("Distribuzione valori di rho per n = 5")
h1.SetFillColorAlpha(9,0.5)
h1.Draw()

c.cd(2)
g1 = ROOT.TGraphErrors(len(N_5), array.array("f", N_5), array.array("f", rho_5),
                      array.array("f", [0]*len(N_5)), array.array("f", rho_err_5))
g2 = ROOT.TGraph(len(N_5), array.array("f", N_5), array.array("f", [rho_mean_5]*len(N_5)))
g1.SetMarkerStyle(2)
g2.SetLineColor(2)
g1.Draw("AP")
g2.Draw("SAME")

c.cd(3)
h2 = Create_histo(rho_10, "rho_val", 15, -1.5,1.5)
h2.SetLineColor(1)
h2.SetTitle("Distribuzione valori di rho per n = 10")
h2.SetFillColorAlpha(9,0.5)
h2.Draw()

c.cd(4)
g3 = ROOT.TGraphErrors(len(N_10), array.array("f", N_10), array.array("f", rho_10),
                      array.array("f", [0]*len(N_10)), array.array("f", rho_err_10))
g4 = ROOT.TGraph(len(N_10), array.array("f", N_10), array.array("f", [rho_mean_10]*len(N_10)))
g3.SetMarkerStyle(2)
g4.SetLineColor(2)
g3.Draw("AP")
g4.Draw("SAME")


c.cd(5)
h3 = Create_histo(rho_15, "rho_val", 20, -1.5,1.5)
h3.SetLineColor(1)
h3.SetTitle("Distribuzione valori di rho per n = 15")
h3.SetFillColorAlpha(9,0.5)
h3.Draw()

c.cd(6)
g5 = ROOT.TGraphErrors(len(N_15), array.array("f", N_15), array.array("f", rho_15),
                      array.array("f", [0]*len(N_15)), array.array("f", rho_err_15))
g6 = ROOT.TGraph(len(N_15), array.array("f", N_15), array.array("f", [rho_mean_15]*len(N_15)))
g5.SetMarkerStyle(2)
g6.SetLineColor(2)
g5.Draw("AP")
g6.Draw("SAME")



c.Draw()
```
