```python
import ROOT
import numpy as np
import pandas as pd
```

```python
fname = "PICO1_run_0000_(1)"
def T(x):
    return ((((x-10000)*3.3)/1.024) -500)/10
```

```python
infile = ROOT.TFile.Open(fname+ ".root")
tree = infile.Get('t1')
leaves = [leave.GetName() for leave in tree.GetListOfLeaves()]
print(len(leaves))
print(leaves)
```

```python
c1 = ROOT.TCanvas('','', 0, 0, 1000, 3000)
c1.Divide(2,8)
for i,leaf in enumerate(leaves):
    c1.cd(i+1)
    tree.Draw(leaf)
c1.Draw()
```

```python
c1 = ROOT.TCanvas('','', 0, 0, 1200, 2000)
c1.Divide(2,4)
chs = ['drift', 'G1T', 'G1B', 'G2T', 'G2B', 'G3T', 'G3B']
for i,ch in enumerate(chs):
    c1.cd(i+1)
    tree.Draw('I_'+ ch + ":V_" + ch, "V_" + ch + ">10000", "COLZ")
c1.Draw()
```

```python
from array import array
nbinsx = 13
minx = 32
maxx = 58
nbinsy = 50
miny = -5
maxy = 5

vi_g1t = ROOT.TH2F('vi_g1t', '; Temperature [#circC] ; Current [nA]', nbinsx, minx, maxx, nbinsy, miny, maxy)
vi_g1b = ROOT.TH2F('vi_g1b', '; Temperature [#circC] ; Current [nA]', nbinsx, minx, maxx, nbinsy, miny, maxy)
vi_g2t = ROOT.TH2F('vi_g2t', '; Temperature [#circC] ; Current [nA]', nbinsx, minx, maxx, nbinsy, miny, maxy)
vi_g2b = ROOT.TH2F('vi_g2b', '; Temperature [#circC] ; Current [nA]', nbinsx, minx, maxx, nbinsy, miny, maxy)
vi_g3t = ROOT.TH2F('vi_g3t', '; Temperature [#circC] ; Current [nA]', nbinsx, minx, maxx, nbinsy, miny, maxy)
vi_g3b = ROOT.TH2F('vi_g3b', '; Temperature [#circC] ; Current [nA]', nbinsx, minx, maxx, nbinsy, miny, maxy)
vi_drift = ROOT.TH2F('vi_drift', '; Temperature [#circC] ; Current [nA]', nbinsx, minx, maxx, nbinsy, miny, maxy)

for i in range(tree.GetEntries()):
    tree.GetEntry(i)
    if(tree.V_G1T>10000):
        vi_g1t.Fill(T(tree.V_G1T), tree.I_G1T)
    if(tree.V_G1B>10000):
        vi_g1b.Fill(T(tree.V_G1B), tree.I_G1B)
    if(tree.V_G2T>10000):
        vi_g2t.Fill(T(tree.V_G2T), tree.I_G2T)
    if(tree.V_G2B>10000):
        vi_g2b.Fill(T(tree.V_G2B), tree.I_G2B)
    if(tree.V_G3T>10000):
        vi_g3t.Fill(T(tree.V_G3T), tree.I_G3T)
    if(tree.V_G3B>10000):
        vi_g3b.Fill(T(tree.V_G3B), tree.I_G3B)
    if(tree.V_drift>10000):
        vi_drift.Fill(T(tree.V_drift), tree.I_drift)
```

```python
f = open("fit_parameters.txt", "w")

c3 = ROOT.TCanvas('','', 0, 0, 1200, 2000)
c3.Divide(2, 4)
c3.Draw()


c3.cd(1)
fg3b = ROOT.TF1("fg3b", "pol1")
vi_g3b.ProfileX().Fit(fg3b, "S")
f.write(str(fg3b.GetParameter(0)) + " " + str(fg3b.GetParameter(1)) + "\n")

c3.cd(2)
fg3t = ROOT.TF1("fg3t", "pol1")
vi_g3t.ProfileX().Fit(fg3t, "S")
f.write(str(fg3t.GetParameter(0)) + " " + str(fg3t.GetParameter(1)) + "\n")


c3.cd(3)
fg2b = ROOT.TF1("fg2b", "pol1")
vi_g2b.ProfileX().Fit(fg2b, "S")
f.write(str(fg2b.GetParameter(0)) + " " + str(fg2b.GetParameter(1)) + "\n")


c3.cd(4)
fg2t = ROOT.TF1("fg2t", "pol1")
vi_g2t.ProfileX().Fit(fg2t, "S")
f.write(str(fg2t.GetParameter(0)) + " " + str(fg2t.GetParameter(1)) + "\n")

c3.cd(5)
fg1b = ROOT.TF1("fg1b", "pol1")
fg1b.SetParameter(0, 7)
vi_g1b.ProfileX().Fit(fg1b, "S")
f.write(str(fg1b.GetParameter(0)) + " " + str(fg1b.GetParameter(1)) + "\n")

c3.cd(6)
fg1t = ROOT.TF1("fg1t", "pol1")
vi_g1t.ProfileX().Fit(fg1t)
f.write(str(fg1t.GetParameter(0)) + " " + str(fg1t.GetParameter(1)) + "\n")

c3.cd(7)
fdrift = ROOT.TF1("fdrift", "pol1")
vi_drift.ProfileX().Fit(fdrift, "S")
f.write(str(fdrift.GetParameter(0)) + " " + str(fdrift.GetParameter(1))) # + "\n"
 
f.close()
```

```python
f = open("fit_parameters.txt", "r")

#print(f.read().split('\n')[0].split(" "))
lines = f.read().split('\n')
#print(lines)
p0g1t = float(lines[5].split(" ")[0])
p1g1t = float(lines[5].split(" ")[1])

p0g1b = float(lines[4].split(" ")[0])
p1g1b = float(lines[4].split(" ")[1])

p0g2t = float(lines[3].split(" ")[0])
p1g2t = float(lines[3].split(" ")[1])

p0g2b = float(lines[2].split(" ")[0])
p1g2b = float(lines[2].split(" ")[1])

p0g3t = float(lines[1].split(" ")[0])
p1g3t = float(lines[1].split(" ")[1])

p0g3b = float(lines[0].split(" ")[0])
p1g3b = float(lines[0].split(" ")[1])

p0drift = float(lines[6].split(" ")[0])
p1drift = float(lines[6].split(" ")[1])

f.close()
```

```python
nbinsx = 100
minx = -1
maxx = 1
i_g1t = ROOT.TH1F('i_g1t', '; Current [nA]', nbinsx, minx, maxx)
i_g1b = ROOT.TH1F('i_g1b', '; Current [nA]', nbinsx, minx, maxx)
i_g2t = ROOT.TH1F('i_g2t', '; Current [nA]', nbinsx, minx, maxx)
i_g2b = ROOT.TH1F('i_g2b', '; Current [nA]', nbinsx, minx, maxx)
i_g3t = ROOT.TH1F('i_g3t', '; Current [nA]', nbinsx, minx, maxx)
i_g3b = ROOT.TH1F('i_g3b', '; Current [nA]', nbinsx, minx, maxx)
i_drift = ROOT.TH1F('i_drift', '; Current [nA]', nbinsx, minx, maxx)
tg1t = 0
tg1b = 0
tg2t = 0
tg2b = 0
tg3t = 0
tg3b = 0
tdrift = 0
for i in range(tree.GetEntries()):
    tree.GetEntry(i)
    if(tree.V_G1T>10000):
        tg1t = T(tree.V_G1T)
    if(tree.V_G1B>10000):
        tg1b = T(tree.V_G1B)
    if(tree.V_G2T>10000):
        tg2t = T(tree.V_G2T)
    if(tree.V_G2B>10000):
        tg2b = T(tree.V_G2B)
    if(tree.V_G3T>10000):
        tg3t = T(tree.V_G3T)
    if(tree.V_G3B>10000):
        tg3b = T(tree.V_G3B)
    if(tree.V_drift>10000):
        tdrift = T(tree.V_drift)
    if(tg1t>0):
        i_g1t.Fill(tree.I_G1T - (p0g1t + p1g1t*tg1t))
    if(tg1b>0):
        i_g1b.Fill(tree.I_G1B - (p0g1b + p1g1b*tg1b))
    if(tg2t>0):
        i_g2t.Fill(tree.I_G2T - (p0g2t + p1g2t*tg2t))
    if(tg2b>0):
        i_g2b.Fill(tree.I_G2B - (p0g2b + p1g2b*tg2b))
    if(tg3t>0):
        i_g3t.Fill(tree.I_G3T - (p0g3t + p1g3t*tg3t))
    if(tg3b>0):
        i_g3b.Fill(tree.I_G3B - (p0g3b + p1g3b*tg3b))
    if(tdrift>0):
        i_drift.Fill(tree.I_drift - (p0drift + p1drift*tdrift))
```

```python
c1 = ROOT.TCanvas("", "", 1000, 1500)
c1.Divide(2,4)
c1.Draw()
c1.cd(1)
i_g1t.Draw("HIST")
c1.cd(2)
i_g1b.Draw("HIST")
c1.cd(3)
i_g2t.Draw("HIST")
c1.cd(4)
i_g2b.Draw("HIST")
c1.cd(5)
i_g3t.Draw("HIST")
c1.cd(6)
i_g3b.Draw("HIST")
c1.cd(7)
i_drift.Draw("HIST")
```
