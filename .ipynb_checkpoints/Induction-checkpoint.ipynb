```python
import ROOT
import pandas as pd
import Functions
import numpy as np
import array
import matplotlib.pyplot as plt
from math import e, sqrt
```

```python
colonne = ["I3B", "I3T", "I2B","I2T", "I1B", "I1T", "ID"]
f1 = ROOT.TF1( "f", "gaus",-60,0)
folder = "arrays/pico_data"
```

```python
Iname_ind = []
for i in range(259,284):
    Iname_ind.append("PICO_I_run0" + str(i)+ ".csv")
```

```python
c = list(np.zeros(7))
j = 0
H = list(np.zeros(7))
ParI_ind = []
SigI_ind = []
Par_err_ind = []
Sig_err_ind = []

for i in colonne:
    c[j] = ROOT.TCanvas("", "", 3000, 6000)
    c[j].Divide(5,7)
    c[j].Draw()
    H[j], parI_ind, sigI_ind, par_err_ind, sig_err_ind =Functions.Do_histo(folder,Iname_ind, f1, i,"I3T", -85, c[j])
    ParI_ind.append(parI_ind)
    SigI_ind.append(sigI_ind)
    Par_err_ind.append(par_err_ind)
    Sig_err_ind.append(sig_err_ind)
    for k in range(len(Iname_ind)):
        H[j][k].SetLineColor(j*j+50)
        H[j][k].Draw()
    j+= 1
```

```python
Vname_ind = []
for i in range(259,284):
    Vname_ind.append("PICO_V_run0" + str(i)+ ".csv")
parV_ind,sigV_ind = Functions.V_mean(folder, Vname_ind, 0)
```

```python
c1= ROOT.TCanvas("","",700,700)
graph = list(np.zeros(7))
j=0
for i in range(7):
    graph[i] = ROOT.TGraph(len(ParI_ind[i]),array.array("f",parV_ind), array.array("f",abs(np.array(ParI_ind[i]))))
    graph[i].SetMarkerStyle(i+2)
    graph[i].SetMarkerColor(i*i + 50)
    if i == 0:
        graph[i].Draw("AP")
        graph[i].GetXaxis().SetTitle("E_{induction} (kV/m)")
        graph[i].GetYaxis().SetTitle("I (nA)")
        graph[i].SetTitle("I vs E_{induction}")
    else:
        graph[i].Draw("SAMEP")
c1.Draw()
#c1.SetLogy()


Leg = ROOT.TLegend(0.1,0.7,0.2,0.9)
Leg.SetHeader("Legend","C")
for i in range(7):
    Leg.AddEntry(graph[i], colonne[i])
Leg.Draw()
```

```python
plt.figure(figsize = (10,10))
for i in range(len(ParI_ind)):
    #plt.scatter(parV_ind, ParI_ind[i])
    plt.errorbar(parV_ind, ParI_ind[i], xerr = sigV_ind, yerr = Par_err_ind[i], linestyle = " ", label = colonne[i], marker ="o")
#plt.yscale("log")
plt.ylabel("$ I (nA)$", size = 16)
plt.xlabel("$E_{induction} (V/mm)$", size =16)
plt.title("$ I \, vs \, E_{induction} $", size =20)
plt.legend()
plt.show()
```
