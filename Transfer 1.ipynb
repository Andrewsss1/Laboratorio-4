```python
colonne = ["I3B", "I3T", "I2B","I2T", "I1B", "I1T", "ID"]
f1 = ROOT.TF1( "f1", "gaus",-60,0)
folder = "arrays/pico_data"
```

```python
c = list(np.zeros(7))
j = 0
H = list(np.zeros(7))
ParI_tr1 = []
SigI_tr1 = []

for i in colonne:
    c[j] = ROOT.TCanvas("", "", 3000, 6000)
    c[j].Divide(5,7)
    c[j].Draw()
    H[j], parI_tr1, sigI_tr1 =Functions.Do_histo(folder,Iname_tr1, f1, i,"I3B", 85, c[j])
    ParI_tr1.append(parI_tr1)
    SigI_tr1.append(sigI_tr1)
    for k in range(len(Iname_tr1)):
        H[j][k].SetLineColor(j*j+50)
        H[j][k].Draw()
    j+= 1
```
